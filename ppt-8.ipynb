```python
#q1
def minimumDeleteSum(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        dp[i][0] = dp[i - 1][0] + ord(s1[i - 1])

    for j in range(1, n + 1):
        dp[0][j] = dp[0][j - 1] + ord(s2[j - 1])

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j] + ord(s1[i - 1]), dp[i][j - 1] + ord(s2[j - 1]))

    return dp[m][n]
```

```python
def isValid(s):
    stack = []

    for char in s:
        if char in '(*':
            stack.append(char)
        elif char == ')':
            if stack and stack[-1] == '(':
                stack.pop()
            elif stack and stack[-1] == '*':
                stack.pop()
            else:
                return False

    while stack:
        if stack[-1] == '(':
            return False
        stack.pop()

    return True
```

```python
def minSteps(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        dp[i][0] = i

    for j in range(1, n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1)

    return dp[m][n]
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def buildTree(s, start):
    if start >= len(s) or s[start] == ')':
        return None

    # Find the value of the current node
    val = 0
    while start < len(s) and s[start] != '(' and s[start] != ')':
        val = val * 10 + int(s[start])
        start += 1

    node = TreeNode(val)

    if start < len(s) and s[start] == '(':
        # Find the left child
        node.left, start = buildTree(s, start + 1)

    if start < len(s) and s[start] == '(':
        # Find the right child
        node.right, start = buildTree(s, start + 1)

    # Move past the closing parenthesis
    if start < len(s) and s[start] == ')':
        start += 1

    return node


def treeToString(root):
    if root is None:
        return ""

    left_str = treeToString(root.left)
    right_str = treeToString(root.right)

    if left_str and right_str:
        return str(root.val) + "(" + left_str + ")(" + right_str + ")"
    elif left_str:
        return str(root.val) + "(" + left_str + ")"
    elif right_str:
        return str(root.val) + "()(" + right_str + ")"
    else:
        return str(root.val)


def constructTreeFromString(s):
    root = buildTree(s, 0)
    return root


def inorderTraversal(root):
    result = []

    def traverse(node):
        if node:
            traverse(node.left)
            result.append(node.val)
            traverse(node.right)

    traverse(root)
    return result
```

```python
def compress(chars):
    write = 0
    read = 0
    count = 0
    prev_char = ""

    while read < len(chars):
        if chars[read] != prev_char:
            if count == 1:
                chars[write] = prev_char
                write += 1
            elif count > 1:
                chars[write] = prev_char
                write += 1
                count_str = str(count)
                for digit in count_str:
                    chars[write] = digit
                    write += 1
            prev_char = chars[read]
            count = 1
        else:
            count += 1
        read += 1

    # Handle the last character and count
    if count == 1:
        chars[write] = prev_char
        write += 1
    elif count > 1:
        chars[write] = prev_char
        write += 1
        count_str = str(count)
        for digit in count_str:
            chars[write] = digit
            write += 1

    return write
```

```python
from collections import Counter


def findAnagrams(s, p):
    p_freq = Counter(p)
    window_freq = Counter()
    left = 0
    right = 0
    match_count = 0
    result = []

    while right < len(s):
        window_freq[s[right]] += 1
        if window_freq[s[right]] == p_freq[s[right]]:
            match_count += 1

        if right - left + 1 > len(p):
            if window_freq[s[left]] == p_freq[s[left]]:
                match_count -= 1
            window_freq[s[left]] -= 1
            if window_freq[s[left]] == 0:
                del window_freq[s[left]]
            left += 1

        if match_count == len(p_freq):
            result.append(left)

        right += 1

    return result
```

```python
def decodeString(s):
    stack = []
    current_string = ""
    current_number = 0

    for char in s:
        if char.isdigit():
            current_number = current_number * 10 + int(char)
        elif char == "[":
            stack.append(current_string)
            stack.append(current_number)
            current_string = ""
            current_number = 0
        elif char == "]":
            num = stack.pop()
            prev_string = stack.pop()
            current_string = prev_string + current_string * num
        else:
            current_string += char

    return current_string
```

```python
def canSwapLetters(s, goal):
    mismatch_count = 0
    mismatch_indices = []

    for i in range(len(s)):
        if s[i] != goal[i]:
            mismatch_count += 1
            mismatch_indices.append(i)

    if mismatch_count == 2:
        i, j = mismatch_indices
        if s[i] == goal[j] and s[j] == goal[i]:
            return True

    return False
```
